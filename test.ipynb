```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data/main_data.xlsx'
data = pd.read_excel(file_path)

data.head(10)
```

```python
for state in data["State"].unique():
    plt.plot(data["Year"].where(data["State"] == state), data["Unemployment %"].where(data["State"] == state))
```

```python
plt.figure(figsize=(14, 10))

for state in data["State"].unique():
    state_data = data[data["State"] == state]
    plt.plot(state_data["Year"], state_data["Unemployment %"] * 100, label=state)

plt.xlabel('Year')
plt.ylabel('Unemployment %')
plt.title('Unemployment Rate by State Over Time')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small', ncol=2)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 10))

for state in data["State"].unique():
    state_data = data[data["State"] == state]
    plt.plot(state_data["Year"], state_data["HDI"], label=state)

plt.xlabel('Year')
plt.ylabel('HDI')
plt.title('HDI by State Over Time')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small', ncol=2)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 10))

for state in data["State"].unique():
    state_data = data[data["State"] == state]
    plt.plot(state_data["Year"], state_data["Gun ownership %"], label=state)

plt.xlabel('Year')
plt.ylabel('Gun ownership %')
plt.title('Gun Ownership % by State Over Time')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small', ncol=2)
plt.grid(True)
plt.tight_layout()
plt.show()
```
