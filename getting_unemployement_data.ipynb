```python
import pandas as pd

input_file_path = 'C:/Users/szczy/Desktop/STUDIA/Magisterka sem1/Data_Analytics/Project/Predicting-USA-presidential-elections/data/ststdsadata.xlsx'
output_file_path = 'C:/Users/szczy/Desktop/STUDIA/Magisterka sem1/Data_Analytics/Project/Predicting-USA-presidential-elections/data/un_new_v7.xlsx'

# Wczytaj plik Excel, pomijając pierwsze 3 wiersze
df = pd.read_excel(input_file_path, skiprows=3)

# Usuń puste wiersze
df.dropna(how='all', inplace=True)

# Ustaw nagłówki ręcznie
new_headers = [
    "FIPS Code", "State and area", "Period", "Month", 
    "Civilian non-institutional population", "Civilian labor force", 
    "Unnamed: 6", "Unnamed: 7", "Unnamed: 8", "Unnamed: 9", "Unemployment"
]
df.columns = new_headers

# Usuń wiersze z nagłówkami kolumn, które nie są potrzebne
df = df[df["FIPS Code"].notna()]

# Wyodrębnij rok i miesiąc z kolumny 'Period'
df['Year'] = df['Period'].astype(str).str[:4]

# Pobierz miesiące z kolumny D od wiersza 9
months_df = pd.read_excel(input_file_path, skiprows=3, usecols="D", header=None)
months = months_df.iloc[8:, 0].values

# Dopasowanie długości kolumny miesięcy do długości głównego dataframe
if len(months) > len(df):
    months = months[:len(df)]
elif len(months) < len(df):
    months = list(months) + [None] * (len(df) - len(months))

# Dodaj miesiące do głównego dataframe
df['Month'] = months

# Upewnij się, że wszystkie wartości w kolumnie 'Month' są wypełnione
df['Month'].fillna(method='ffill', inplace=True)

# Zmień nazwę kolumny zawierającej bezrobocie
df.rename(columns={'Unemployment': 'Unnamed: 10'}, inplace=True)

# Wybierz odpowiednie kolumny
selected_columns = df[['State and area', 'Year', 'Month', 'Unnamed: 10']]

# Zapisz dane do nowego pliku Excel
selected_columns.to_excel(output_file_path, index=False)

# Potwierdzenie zapisu pliku
print("Plik został zapisany jako 'unemployment_data_new_v7.xlsx'")
```

```python
import pandas as pd

# Wczytaj dane
input_file_path = 'C:/Users/szczy/Desktop/STUDIA/Magisterka sem1/Data_Analytics/Project/Predicting-USA-presidential-elections/data/un_new_v7.xlsx'
df = pd.read_excel(input_file_path)

# Sprawdź dostępne lata w danych
available_years = df['Year'].unique()
print("Available years in the data:", available_years)

# Lista stanów w danych
states = df['State and area'].unique()
print("States in the data:", states)

# Wskaż lata, dla których chcesz obliczyć średnią
years = [1992, 1996, 2000, 2004, 2008, 2012, 2016, 2020]

# Usuń spacje i inne zbędne znaki z kolumny 'Year'
df['Year'] = df['Year'].astype(str).str.strip()

# Lista do przechowywania wyników
all_averages = []

# Iteruj przez każdy stan i oblicz średnią roczną stopę bezrobocia dla wskazanych lat
for state in states:
    state_df = df[df['State and area'] == state]
    for year in years:
        year_str = str(year)  # Konwertuj rok na string
        year_data = state_df[state_df['Year'] == year_str]
        if not year_data.empty:
            # Zakładamy, że właściwa kolumna nazywa się 'Unemployement'
            year_data['Unemployement'] = year_data['Unemployement'].astype(str).str.replace(',', '.').astype(float)
            average_unemployment = round(year_data['Unemployement'].mean(), 3)
            all_averages.append((state, year, average_unemployment))
        else:
            all_averages.append((state, year, None))

# Przekształć wyniki do dataframe
all_averages_df = pd.DataFrame(all_averages, columns=['State', 'Year', 'Average Unemployment'])

# Zapisz wyniki do nowego pliku Excel
output_file_path = 'C:/Users/szczy/Desktop/STUDIA/Magisterka sem1/Data_Analytics/Project/Predicting-USA-presidential-elections/data/all_states_unemployment_averages.xlsx'
all_averages_df.to_excel(output_file_path, index=False)

# Potwierdzenie zapisu pliku
print("Plik został zapisany jako 'all_states_unemployment_averages.xlsx'")
```
